```python
import graph_tool.all as gt
```

```python
g = gt.Graph()
```

```python
v1 = g.add_vertex()
v2 = g.add_vertex()
```

```python
e = g.add_edge(v1,v2)
```

```python
gt.graph_draw(g,
    vertex_text = g.vertex_index,
    vertex_font_size = 18,
    output_size =(200,200)
)
```

```python
g = gt.lattice([30,20] , periodic = True)
pos = gt.sfdp_layout(g, cooling_step=0.95,epsilon=1e-2)
gt.graph_draw(g , pos = pos,output_size=(300,300))
```

```python
g = gt.complete_graph(20)
pos = gt.sfdp_layout(g,cooling_step=0.95,epsilon=1e-2)
gt.graph_draw(g,pos=pos,output_size=(300,300))
```

```python
g = gt.circular_graph(30,3)
pos = gt.sfdp_layout(g , cooling_step=0.95,epsilon=1e-2)
gt.graph_draw(g,pos=pos,output_size=(200,200))
```
