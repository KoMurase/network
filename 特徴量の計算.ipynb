```
import matplotlib.pyplot as plt 
import networkx as nx 

rnd = nx.gnp_random_graph(10, 0.3)
plt.subplot(331)
nx.draw_circular(rnd, node_size=200, node_color='red', with_labels=True,font_weight='bold')

petersen = nx.petersen_graph()
plt.subplot(332)
nx.draw_shell(petersen,nlist=[range(5,10), range(5)],node_size=200, node_color='blue', with_labels=True, font_weight='bold')

cycle=nx.cycle_graph(10)
plt.subplot(333)
nx.draw_circular(cycle, node_size=200, node_color='yellow', with_labels=True, font_weight='bold')

K_10= nx.complete_graph(10)
plt.subplot(334)
nx.draw_circular(K_10, node_size=200,node_color='green', with_labels=True, font_weight='bold')

K_5_5= nx.complete_bipartite_graph(5, 5)
plt.subplot(335)
nx.draw_circular(K_5_5, n_list=[range(5,10), range(5)],node_size=200, node_color='pink', with_labels=True, font_weight='bold')

barbell = nx.barbell_graph(4, 2)
plt.subplot(336)
nx.draw_spring(barbell, node_size=200, node_color='purple', with_labels=True, font_weight='bold')

star = nx.star_graph(9)
plt.subplot(337)
nx.draw(star, node_size=200, node_color='red', with_labels=True, font_weight='bold')

wheel = nx.wheel_graph(10)
plt.subplot(338)
nx.draw(wheel, node_size=200, node_color='blue', with_labels=True, font_weight='bold')

ba = nx.barabasi_albert_graph(10,2)
plt.subplot(339)
nx.draw(ba, node_size=200, node_color='green', with_labels=True, font_weight='bold')
```

頂点の数や辺の数などネットワークの構造的特徴を表す量を特徴量という<br>
頂点の数や辺の数が同じでも上のように構造が異なるネットワークが考えられる
