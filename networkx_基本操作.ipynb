```
import networkx as nx 
G = nx.Graph()  #空のネットワークGを作る
G.add_nodes_from(["A","B","C","D"]) #Gにノードを追加する　
G.add_edges_from([("A", "B"),("B", "C"),("C", "D"),("D", "A")]) #("A", "B")とするとAB間に辺を追加する #数値でも可
nx.draw(G, node_size=400, node_color='yellow', with_labels=True,font_weight='bold') #Gの描画
```
