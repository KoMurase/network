```
import networkx as nx 
from google.colab import files
uploaded = files.upload()
```

```

for fn in uploaded.keys():
  print('アップロードファイル "{name}" lengh{length}bytes'.format(name=fn, length=len(uploaded[fn])))

  G = nx.readwrite.gml.read_gml(fn)
  nx.draw_spring(G, node_size=200,node_color='red',width_labels=True)
```
